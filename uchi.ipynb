```python
#Nama: Suchi Ramadhani
#NIM : 2507046003

import numpy as np       # numpy berfungsi untuk operasi matematika berbasis array,
                         # memungkinkan perhitungan cepat pada data besar.

import matplotlib.pyplot as plt

# PARAMETER DASAR FLUIDA (TEKANAN HIDROSTATIS)
# Kedalaman dibuat menggunakan np.linspace untuk menghasilkan titik data yang rapat.

rho = 100                 # massa jenis "fluida ringan" (kg/m^3)  lebih kecil agar tekanannya tidak besar
g = 9.8                   # percepatan gravitasi (m/s^2)
h = np.linspace(0, 5, 500)   # kedalaman dari 0 sampai 5 meter, 500 titik
                               # lebih sedikit titik dan kedalaman agar hasil tidak terlalu besar

# PROSES 1  Tekanan Hidrostatis
P1 = rho * g * h           # np array: menghitung  g h untuk semua elemen sekaligus

# PROSES 2  Gaya Hidrostatis pada Luas Bidang
A = 1.0                    # luas bidang (m^2)  lebih kecil agar gaya tidak terlalu besar
P2 = P1 * A

# PROSES 3  Superposisi Tekanan
alpha = 0.2                # faktor koreksi eksponensial
P3 = P1 + (P1 * np.exp(-alpha * h) * 0.3)   # np.exp: menghitung e^x untuk seluruh array

# PERHITUNGAN NILAI HASIL
hasil = (
    "HASIL PERHITUNGAN PROGRAM\n"
    "--------------------------------------------\n"
    f" Tekanan Maks Hidrostatis       : {np.max(np.abs(P1)):.2f} Pa\n"
    f" Gaya Hidrostatis Maks          : {np.max(np.abs(P2)):.2f} N\n"
    f" Tekanan Maks Superposisi       : {np.max(np.abs(P3)):.2f} Pa\n"
    "\n"
    f" Kedalaman Maks                 : {np.max(h):.2f} meter\n"
    f" Luas Bidang                    : {A} m^2\n"
    f" Faktor Koreksi Eksponensial    : {alpha}\n"
    "\n"
    f" Jumlah Titik Perhitungan       : {len(h)} titik\n"
)

# PLOTTING

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)

plt.plot(h, P1, label="Tekanan Hidrostatis (P1)")
plt.plot(h, P2, label="Gaya Hidrostatis (P2)")
plt.plot(h, P3, label="Superposisi Tekanan (P3)")

plt.title("Grafik Tekanan dan Gaya Hidrostatis pada Fluida (Nilai Lebih Kecil)")
plt.xlabel("Kedalaman (m)")
plt.ylabel("Tekanan / Gaya")
plt.grid(True)
plt.legend()

# AREA TEKS DI BAWAH GRAFIK

plt.subplot(2, 1, 2)
plt.axis("off")
plt.text(0, 1, hasil, fontsize=10, family='monospace', va='top')

plt.tight_layout()
plt.show()
```
